```python
def read_from_file(dire):
    file_ = open(dire)
    time = []
    temp = []
    for l in file_.readlines():
        if l[0] != '#' and l[0] != '@':
            l = l.split(' ')
            li=[]
            for c in l:
                if c != '':
                    li.append(c)
            time.append(li[0])
            temp.append(li[1])
    file_.close()
    return time, temp
```

```python
def read_from_file_new(dire):
    file_ = open(dire)
    time = []
    temp = []
    temp2 = []
    temp3 = []
    for l in file_.readlines():
        if l[0] != '#' and l[0] != '@':
            l = l.split(' ')
            li=[]
            for c in l:
                if c != '':
                    li.append(c)
            time.append(li[0])
            temp.append(li[1])
            temp2.append(li[2])
            temp3.append(li[3])
    file_.close()
    return time, temp, temp2, temp3
```

```python
#temperatur plot
def argon_temp():
    arg_dirs= dir_argon[:5]
    for dir_arg in arg_dirs:
        time, temp = read_from_file(dir_arg+str_temp)
        fig = plt.figure()
        ax = fig.add_subplot(111)
        ax.set_xlabel('Time [ps]')
        ax.set_ylabel('Temperature [K]')
        ax.set_title('Temperature')
        ax.plot(time, temp)
        plt.savefig('temp_cool_'+dir_arg[-5:-2]+'.pdf')
```

```python
def sigmoid_function(x, L, k, x0, y0):
    return L/(1+np.exp(-k*(x-x0))) + y0
```

```python
#potential plot
prms = [[-9.43231934e+02,1.01768680e-02,1.15371489e+03,-1.09417771e+01],
        [ -5.18704911e+02 , 3.50428968e-03,3.98964101e+03,-1.13009313e+01],
        [ -5.14874952e+02,3.28154635e-03,6.23966032e+03,-1.11637715e+01],
        [5000.,1e-3,6000,-1.],
        [-5000.,-1e-2,-1000,-1.]]
#-5.14874952e+02   3.28154635e-03   6.23966032e+03  -1.11637715e+01
def argon_pot():
    arg_dirs= dir_argon[:3]
    for j, dir_arg in enumerate(arg_dirs):
        time,pot = None,None
        if dir_arg == dir_argon[4]:
            time,pot = read_from_file(dir_arg+'energy_pot.xvg')
        else:
            time,pot = read_from_file(dir_arg+str_pot)
        
        #params = prms[j]
        params = prms[3]
        time,pot = np.array(time).astype(float),np.array(pot).astype(float)
        #logistic regression
        params , cov = curve_fit(sigmoid_function, time, pot, prms[3])
        print dir_arg, params, cov
        fitted = []
        for t in time:
            fitted.append(sigmoid_function(t, params[0], params[1], 
                          params[2], params[3]))
        fitted = np.array(fitted)
        
        
        fig = plt.figure()
        ax = fig.add_subplot(111)
        ax.set_xlabel('Time [ps]')
        ax.set_ylabel('Potential Energy [kJ/mol]')
        ax.set_title('Potential Energy')
        ax.plot(time, pot, label='Data')
        ax.plot(time,fitted, color='red', label='Fit')
        ax.legend(loc=3)
        
        #fig2 = plt.figure()
        #ax = fig.add_subplot(111)
        #fitted = []
        #for i in np.arange(0,10000,1):
        #    fitted.append(sigmoid_function(i,params[0], params[1], 
        #                  params[2], params[3]))
        #ax.plot(np.arange(0,10000,1),fitted, color='red')

        plt.savefig('pot_cool_'+dir_arg[-5:-2]+'.pdf')
        
```

```python
def cool_vs_heatup():
    time, temp_cool = read_from_file(dir_argon[2]+str_temp)
    time, temp_heat = read_from_file(dir_argon[3]+str_temp)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xlabel('Time [ps]')
    ax.set_ylabel('Temperature [K]')
    ax.set_title('Temperature')
    ax.plot(time, temp_cool, color='blue', label = 'Cooling')
    ax.plot(time, temp_heat, color='red', label = 'Heating')
    ax.legend(loc=2)
    plt.savefig('heatup_temp.pdf')
    
    time, pot_cool = read_from_file(dir_argon[2]+str_pot)
    time, pot_heat = read_from_file(dir_argon[3]+str_pot)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xlabel('Time [ps]')
    ax.set_ylabel('Potential [kJ/mol]')
    ax.set_title('Potential')
    ax.plot(time, pot_cool, color='blue', label = 'Cooling')
    ax.plot(time, pot_heat, color='red', label = 'Heating')
    
    time,pot_cool = np.array(time).astype(float),np.array(pot_cool).astype(float)

    fitted = []
    params = prms[2]
    for t in time:
        fitted.append(sigmoid_function(t, params[0], params[1], 
                      params[2], params[3]))
    fitted = np.array(fitted)
    ax.plot(time,fitted, color='green', lw=2.5, label='Fit')
    ax.legend(loc=5)
    plt.savefig('heatup_pot.pdf')

```

```python
def pca():
    number, eigenval = read_from_file(dir_argon[6]+'eigenval.xvg')
    eigenval = np.array(eigenval).astype(float)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xlabel('Principle Component')
    ax.set_ylabel('Variance [nm$^2$]')
    ax.set_title('Variance of the Principle Components')
    ax.plot(number[:818], eigenval[:818])
    ax.set_yscale('log')
    ax.set_xlim(0,818)
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xlabel('Principle Component')
    ax.set_ylabel('Variance [nm$^2$]')
    ax.set_title('Variance of the Principle Components')
    ax.plot(number[818:], -eigenval[818:])
    ax.set_yscale('log')
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xlabel('Principle Component')
    ax.set_ylabel('% of Variance expained')
    ax.set_title('Variance of the Principle Components')
    var_tot = np.sum(eigenval)
    var_cum = []
    for i in np.arange(0,20,1):
        var_cum.append(np.sum(eigenval[:i+1])/var_tot*100)
    var_cum = np.array(var_cum)
    ax.grid(True)
    ax.bar(np.arange(0,20,1),var_cum, label='Cumulative Variance', color='salmon')
    ax.bar(np.arange(0,20,1),eigenval[:20]/var_tot*100, label='Variance')
    print var_cum[1]
    ax.legend(loc=5)
    plt.savefig('pca_comp.pdf')
    #ax.set_yscale('log')
```

```python
def read_from_file_new_new(dire):
    file_ = open(dire)
    time = [[],[],[],[],[],[],[],[]]
    temp = [[],[],[],[],[],[],[],[]]
    point_arr = []
    point = 0
    rl = file_.readlines()

    
    for l in rl:
        if l[0]=='&':
            point+=1
        if l[0] != '#' and l[0] != '@' and l[0] != '&':
            l = l.split(' ')
            li=[]
            for c in l:
                if c != '':
                    li.append(c)
            time[point].append(li[0])
            temp[point].append(li[1])
    file_.close()
    return time, temp
```
