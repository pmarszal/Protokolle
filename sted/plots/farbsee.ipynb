```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib 
%matplotlib inline
import scipy as sp
from scipy.optimize import curve_fit
from scipy import special
import sympy as smp
def read_ff(fn):
    f= open(fn)
    ls = f.readlines()
    x,y = [],[]
    for l in ls:
        l = l.split('\t')
        x.append(l[0]), y.append(l[1])
    return np.array(x).astype(float), np.array(y).astype(float)
def read_data(fn):
    f= open(fn)
    ls = f.readlines()
    x = []
    for l in ls:
        l = l.split(',')
        x.append(np.array(l).astype(float))
    return np.array(x).astype(float)
def read_fwhm(FN):
    x = []
    for fn in FN:
        y = []
        f= open(fn)
        fnew = fn.split('/')[-1]
        fnew = fnew.split('_sted')[0]
        fnew = fnew.replace('_','.')
        y.append(float(fnew))
        ls = f.readlines()
        
        for l in ls:
            if l[:6] == '%Width':
                #print l
                l = l.split('\t')
                y.append(l[-1])
        x.append(np.array(y).astype(float))
    return np.array(x).astype(float)
def arr_2_table(arr):
    no_row = len(arr)
    no_col = len(arr[0])//2+1
    head = '''
% Python latex table
\\begin{tabular}'''
    head_arg = '{l|'
    for i in range(no_col):
        head_arg = head_arg + 'c'
    head_arg = head_arg + '}'
    head = head + head_arg + '\n'
    hlabels = ''
    for i in range(no_col):
        hlabels = hlabels+'&'
    hlabels = hlabels+' \\\\ \\hline \n'
    table_content = hlabels
    for i in range(no_row):
        row = arr[i]
        strrow = row[0]
        j = 1
        while j < len(row):
            strrow = strrow+'  &  '
            val = row[j]
            err = row[j+1]
            if err < 1:
                counter = 0
                while err <1:
                    err = err*10.
                    val = val*10
                    counter +=1
                val = np.round(val)
                err = np.ceil(err)
                while counter >0:
                    val = val/10.
                    counter -=1
            
            else:
                counter = 0
                while err >1:
                    err = err/10.
                    val = val/10.
                    counter +=1
                val = np.round(val)
                err = np.ceil(err)
                while counter >0:
                    err = err*10.
                    val = val*10
                    counter -=1
            strrow =strrow + str(val)+'('+str(err)+')'
            j=j+2
        strrow = strrow + '\\\\ \n'
        table_content = table_content+strrow
    foot = '''
\end{tabular}
'''
    return head + table_content + foot
```
