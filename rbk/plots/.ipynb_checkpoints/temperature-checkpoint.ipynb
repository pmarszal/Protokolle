```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import scipy as sp
from scipy.optimize import curve_fit
from matplotlib import gridspec
import pickle
%matplotlib inline

#plt.rcParams.update({'axes.titlesize': 'medium'})
#plt.rcParams.update({'axes.labelsize': 'small'})
plt.rcParams['font.size']= 15


fit_color = 'black'
data_color = 'red'
data_color_sec = 'blue'
fit_width = 1.5

D_T = np.absolute(11.049884-19.836957)
g = 9.81 # m/s^2
alpha = 2.1*10**(-4) #1/K
rho = 1000. # kg/m^3
c_p = 4187 # J/kg/K
mu = 1*10.**(-3) # Pa*s
Lambda = 0.6 # W/m/K
kappa = Lambda/(rho*c_p) # 
nu = mu/rho #
print nu
L = 0.2 # m
Ray_exp = (g*alpha*D_T*L**3)/(kappa*nu)
print Ray_exp
```

# General Functions

```python
def check_empty(arr):
    empty_list = []
    for i, e in enumerate(arr):
        if e != '' and e != '\n':
            empty_list.append(i)
    return empty_list
```

```python
# File Reading GENERALIZED!!! YEEEEEEEEEAAAAAAAAAAAAAAHH!
def read_from_file(filename, row = 'all', col='all', separator = '\t', comment='#'):
    f_in = open(filename)
    table = []
    lines = f_in.readlines()
    lines = np.array(lines)
    rows = None
    no_col = 0 
    if row == 'all':
        rows = range(len(lines))
    else:
        rows = row
    for line in lines[rows]:
        if line[0]!='#':
            l=np.array(line.split(separator))
            empty_list = check_empty(l)
            l = l[empty_list]
            if len(l)>no_col:
                no_col = len(l)
            if col == 'all':
                l = np.array(l).astype(float)
            else:
                l_new=[]
                for i in col:
                    l_new.append(l[i])
                l = np.array(l_new).astype(float)
            table.append(l)
    #Fill with zeros at the end
    for i in range(len(table)):
        table[i] = list(table[i])
        while len(table[i])<no_col:
            table[i].append('0')
    table = np.array(table).astype(float)
    return table
```
